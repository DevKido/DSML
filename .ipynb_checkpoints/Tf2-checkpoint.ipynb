```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv("./tf2_data/DATA/fake_reg.csv")
```

```python
data.head()
```

```python
X = data[["feature1","feature2"]].values
```

```python
y= data["price"].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
%matplotlib inline
losses.plot()
```

```python
%%time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
%%time
data = pd.read_csv("./tf2_data/DATA/cancer_classification.csv")
```

```python
data.head()
```

```python
sns.countplot("benign_0__mal_1", data=data)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(data.corr())
```

```python
X = data.drop("benign_0__mal_1", axis=1).values
y = data["benign_0__mal_1"].values
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = Sequential()
model.add(Dense(30,activation="relu"))
model.add(Dense(15,activation="relu"))
model.add(Dense(1,activation="sigmoid"))
model.compile(optimizer="adam",loss="binary_crossentropy")
```

```python
pd.DataFrame(model.history.history).plot()
```

```python
model = Sequential()
model.add(Dense(30,activation="relu"))
model.add(Dense(15,activation="relu"))
model.add(Dense(1,activation="sigmoid"))
model.compile(optimizer="adam",loss="binary_crossentropy")
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
pd.DataFrame(model.history.history).plot()
```

```python
model = Sequential()
model.add(Dense(30,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(15,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(1,activation="sigmoid"))
model.compile(optimizer="adam",loss="binary_crossentropy")
```

```python
model.fit(X_train,y_train, validation_data=(X_test,y_test), epochs=600, callbacks=[ea])
```

```python
pd.DataFrame(model.history.history).plot()
```

```python
preds = model.predict_classes(X_test)
```

```python
print(classification_report(y_test,preds))
```

```python
print(confusion_matrix(y_test,preds))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report, confusion_matrix
```

```python
from tensorflow.keras.models import Sequential,load_model
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
data = pd.read_csv("./tf2_data/DATA/lending_club_loan_two.csv")
```

```python
data.describe().transpose()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.drop("emp_title",axis=1,inplace=True)
```

```python
sns.countplot(x="emp_length",data=data,hue="loan_status")
```

```python
100 * data[data["loan_status"]!="Fully Paid"].groupby("emp_length").count()["loan_status"] / (data[data["loan_status"]=="Fully Paid"].groupby("emp_length").count()["loan_status"] + data[data["loan_status"]!="Fully Paid"].groupby("emp_length").count()["loan_status"] )
```

```python
data.drop("emp_length",axis=1,inplace=True)
```

```python
data["target"] = data["loan_status"].map({"Fully Paid":1, "Charged Off": 0})
```

```python
data.isnull().sum()
```

```python
data.drop("title",axis=1,inplace=True)
```

```python
data.corr()["mort_acc"].sort_values()
```

```python
mort_lookup = data.groupby("total_acc").mean()["mort_acc"]

def fill_mort(tot_acc, mort_acc):
    if np.isnan(mort_acc):
        return mort_lookup[tot_acc]
    else:
        return mort_acc
```

```python
data["mort_acc"] = data.apply(lambda x : fill_mort(x["total_acc"],x["mort_acc"]),axis=1)
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
data.select_dtypes("object").columns
```

```python
data["term"].value_counts()
```

```python
data["term"] = data["term"].apply(lambda x : int(x[:3]))
```

```python
data.drop("grade",axis=1,inplace=True)
```

```python
data["home_ownership"] = data["home_ownership"].replace(["NONE","ANY"],"OTHER")
```

```python
data["earliest_cr_line"] = data["earliest_cr_line"].apply(lambda x : int(x[-4:]))
```

```python
data.drop("issue_d",axis=1,inplace=True)
```

```python
data["zipcode"] = data["address"].apply(lambda zip: int(zip[-5:]))
```

```python
data["zipcode"] = data["zipcode"].apply(lambda zip: str(zip))
```

```python
data.drop("address",axis=1,inplace=True)
```

```python
dummies = pd.get_dummies(data[["home_ownership","zipcode","sub_grade","verification_status","purpose","initial_list_status","application_type"]],drop_first=True)
```

```python
final_data = pd.concat([data,dummies],axis=1)
```

```python
final_data.drop("loan_status",axis=1,inplace=True)
```

```python
final_data.select_dtypes("object").columns
```

```python
final_data.drop(['sub_grade', 'home_ownership', 'verification_status', 'purpose',
       'initial_list_status', 'application_type', 'zipcode'],axis=1,inplace=True)
```

```python
X = final_data.drop("target",axis=1).values
y = final_data["target"].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=101)
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = Sequential()

model.add(Dense(78,activation="relu"))
model.add(Dropout(0.2))


model.add(Dense(39,activation="relu"))
model.add(Dropout(0.2))

model.add(Dense(19,activation="relu"))
model.add(Dropout(0.2))

model.add(Dense(1,activation='sigmoid'))

model.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
ea = EarlyStopping(monitor="val_loss", mode="min",patience = 10, verbose=1)
```

```python
model.fit(X_train,y_train, validation_data=(X_test,y_test), batch_size=512, epochs=250, callbacks=[ea])
```

```python
pd.DataFrame(model.history.history).plot(figsize=(12,8))
```

```python
preds = model.predict_classes(X_test)
```

```python
print(classification_report(y_test,preds))
print("\n\n\n")
print(confusion_matrix(y_test,preds))
```

```python
model.save("./loan_payback.h5")
```
